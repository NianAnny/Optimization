```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 5
def objective(x):
    return x**2 + 1

# Define the Lagrangian
def lagrangian(x, lam):
    return objective(x) + lam * (x - 2) * (x - 4)

# Define the range for x
x = np.linspace(0, 5, 400)
y = objective(x)

# Plot the objective
plt.plot(x, y, label='$x^2 + 1$')

# Plot the Lagrangian for several values of lambda
lambdas = [0.5, 1, 2, 3]  # lambda values >= 0
for lam in lambdas:
    plt.plot(x, lagrangian(x, lam), label=f'Lagrangian with $\lambda={lam}$')
    
plt.xlabel('$x$')
plt.ylabel('$y$ and $L(x, \lambda)$')
plt.legend()
plt.show()
```

```python
# Problem 6
import cvxpy as cp 

x = cp.Variable(2)
obj = cp.Minimize(x[0]**2 + x[1]**2)
constraints = [(x[0] - 1)**2 + (x[1] - 1)**2 <= 1,
               (x[0] - 1)**2 + (x[1] + 1)**2 <= 4]
prob = cp.Problem(obj, constraints)
result = prob.solve()
print('Optimal value:', result)
print(x.value)
```
